```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

sns.set()

def plota_pivot_table(df, value, index, func, ylabel, xlabel, opcao='nada'):
    if opcao == 'nada':
        pd.pivot_table(df, values=value, index=index,aggfunc=func).plot(figsize=[15, 5])
    elif opcao == 'unstack':
        pd.pivot_table(df, values=value, index=index,aggfunc=func).unstack().plot(figsize=[15, 5])
    elif opcao == 'sort':
        pd.pivot_table(df, values=value, index=index,aggfunc=func).sort_values(by=value).plot(figsize=[15, 5])

    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    plt.show()

# Substitua os caminhos dos arquivos, as colunas e as funes de agregao conforme necessrio
meses = ['MAR', 'ABR', 'MAI', 'JUN', 'DEZ']
for mes in meses:
    df = pd.read_csv(f'drive/MyDrive/input/SINASC_RO_2019_{mes}.csv')
    plota_pivot_table(df, 'PESO', 'IDADEMAE', 'mean', 'Peso Mdio', f'Idade da Me ({mes})')

```

Utilizar um loop para percorrer os nomes dos arquivos de cada ms e gerar grficos automaticamente.

Criar uma funo que aceite o nome do arquivo e as colunas a serem analisadas como parmetros.


---

**Funo para Gerar Grficos Baseada no Nome do Arquivo e Colunas**

```python
def gerar_grafico_por_mes(arquivo, valor_coluna, indice_coluna, func_agregacao):
    # Ler o arquivo CSV
    df = pd.read_csv(arquivo)

    # Gerar o grfico usando a funo `plota_pivot_table`
    plota_pivot_table(df, valor_coluna, indice_coluna, func_agregacao, f'{func_agregacao.capitalize()} de {valor_coluna}', indice_coluna)
```

```python
gerar_grafico_por_mes('drive/MyDrive/input/SINASC_RO_2019_ABR.csv', 'PESO', 'IDADEMAE', 'mean')
```

# Parmetros:
arquivo: O caminho para o arquivo CSV que voc deseja analisar.

valor_coluna: A coluna do DataFrame que voc quer analisar (por exemplo, 'PESO').

indice_coluna: A coluna que servir como ndice no grfico (por exemplo, 'IDADEMAE').

func_agregacao: A funo de agregao que voc deseja usar (por exemplo, 'mean' para mdia).

---

**Loop para Percorrer os Nomes dos Arquivos e Gerar Grficos**

```python
# Lista de meses para anlise
meses = ['MAR', 'ABR', 'MAI', 'JUN', 'DEZ']

# Loop para percorrer cada ms e gerar o grfico
for mes in meses:
    arquivo = f'drive/MyDrive/input/SINASC_RO_2019_{mes}.csv'
    gerar_grafico_por_mes(arquivo, 'PESO', 'IDADEMAE', 'mean')
```
