```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

sns.set()

def plota_pivot_table(df, value, index, func, ylabel, xlabel, opcao='nada'):
    if opcao == 'nada':
        pd.pivot_table(df, values=value, index=index,aggfunc=func).plot(figsize=[15, 5])
    elif opcao == 'unstack':
        pd.pivot_table(df, values=value, index=index,aggfunc=func).unstack().plot(figsize=[15, 5])
    elif opcao == 'sort':
        pd.pivot_table(df, values=value, index=index,aggfunc=func).sort_values(by=value).plot(figsize=[15, 5])

    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    plt.show()

# Substitua os caminhos dos arquivos, as colunas e as funções de agregação conforme necessário
meses = ['MAR', 'ABR', 'MAI', 'JUN', 'DEZ']
for mes in meses:
    df = pd.read_csv(f'drive/MyDrive/input/SINASC_RO_2019_{mes}.csv')
    plota_pivot_table(df, 'PESO', 'IDADEMAE', 'mean', 'Peso Médio', f'Idade da Mãe ({mes})')

```

#  Automação
Utilizar um loop para percorrer os nomes dos arquivos de cada mês e gerar gráficos automaticamente.

Criar uma função que aceite o nome do arquivo e as colunas a serem analisadas como parâmetros.


---

**Função para Gerar Gráficos Baseada no Nome do Arquivo e Colunas**

```python
def gerar_grafico_por_mes(arquivo, valor_coluna, indice_coluna, func_agregacao):
    # Ler o arquivo CSV
    df = pd.read_csv(arquivo)

    # Gerar o gráfico usando a função `plota_pivot_table`
    plota_pivot_table(df, valor_coluna, indice_coluna, func_agregacao, f'{func_agregacao.capitalize()} de {valor_coluna}', indice_coluna)
```

```python
gerar_grafico_por_mes('drive/MyDrive/input/SINASC_RO_2019_ABR.csv', 'PESO', 'IDADEMAE', 'mean')
```

# Parâmetros:
arquivo: O caminho para o arquivo CSV que você deseja analisar.

valor_coluna: A coluna do DataFrame que você quer analisar (por exemplo, 'PESO').

indice_coluna: A coluna que servirá como índice no gráfico (por exemplo, 'IDADEMAE').

func_agregacao: A função de agregação que você deseja usar (por exemplo, 'mean' para média).

---

**Loop para Percorrer os Nomes dos Arquivos e Gerar Gráficos**

```python
# Lista de meses para análise
meses = ['MAR', 'ABR', 'MAI', 'JUN', 'DEZ']

# Loop para percorrer cada mês e gerar o gráfico
for mes in meses:
    arquivo = f'drive/MyDrive/input/SINASC_RO_2019_{mes}.csv'
    gerar_grafico_por_mes(arquivo, 'PESO', 'IDADEMAE', 'mean')
```
